# Manipulation et visualisation des images

## liste des répertoires

```python
import os  # pour acceder aus repertoires et fichiers
import cv2  # open-cv bibliothèque de traitement d'image
import matplotlib.pyplot as plt 

Categories = os.listdir('Images')
```

## Liste des fichiers d'un répertoire

```python
Files0 = os.listdir('Images/'+Categories[0])
```

```python
for f in Files0[:5] :
    im = cv2.imread('Images/'+Categories[0]+'/'+f)
    print(im.shape)
```

## Affichage - Crop - Blank-fill - Resizing 

### Affichage

```python
plt.imshow(im)
```

### Crop

```python
im2 = im[50:250,50:350]
plt.imshow(im2)
```

### Blank - Fill

```python
import numpy as np
im3 = np.zeros((768,1024,3),np.uint8)
im3[200:200+im.shape[0],300:300+im.shape[1]]=im[:,:]
plt.imshow(im3)
```

### Resize

```python
im4= cv2.resize(im, (128, 128),interpolation=cv2.INTER_NEAREST)
plt.imshow(im4)
```
